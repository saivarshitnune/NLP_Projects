### Table extraction from PDF
- Now let’s say that your PDF has tables and let’s say you want to preserve the structure of the tables.
- You will have to specify the [strategy](https://unstructured-io.github.io/unstructured/best_practices/strategies.html) parameter as `hi_res`. This will use a combination of computer vision and Optical Character Recognition (OCR) to extract the tables and maintain the structure.
It will return both the text and the html of the table. This is super useful for rendering the tables or passing to a LLM.

> Note: For even better table extraction Unstructured offers an API that improves upon the existing open source models.

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained("mistralai/Mistral-7B-Instruct-v0.2")

# Load quantized model
model = AutoModelForCausalLM.from_pretrained(
    "mistralai/Mistral-7B-Instruct-v0.2",
    device_map="auto",  # Automatically map layers to GPUs
    load_in_8bit=True,  # Enable 8-bit quantization
    torch_dtype=torch.float16  # Use FP16 precision for speed and memory efficiency
)

# Initialize pipeline
llm_pipeline = pipeline("text-generation", model=model, tokenizer=tokenizer)

# # Example usage
# output = llm_pipeline("What is quantization?", max_length=50)
# print(output)
```

```python
# Convert HTML table to pandas DataFrame
import pandas as pd
dfs = pd.read_html(table_html)
df = dfs[0]  # Assuming the table is the first one
print(df.head())
```

```python
table_text = df.to_string(index=False)  # Convert DataFrame to string without index
print(f"Table Text:\n{table_text}")
```

```python
# Prompt for table summarization
prompt = (
    "Summarize the following table data:\n\n"
    f"{table_text}\n\n"
    "Provide a concise summary of the key points."
)

# Generate summary using max_new_tokens instead of max_length
response = llm_pipeline(prompt, max_new_tokens=100, num_return_sequences=1)
summary = response[0]['generated_text']
print(f"Summary:\n{summary}")
```
